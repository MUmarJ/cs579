```python
import tweepy
import json
import csv
from textblob import TextBlob
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
plt.style.use('seaborn-talk')
```

```python
with open('login.json') as json_file:
    twitter_creds = json.load(json_file)
```

```python
bearer_token = twitter_creds["login"]["bearer_token"]
```

```python
##### Pineapple on Pizza sentiment
client = tweepy.Client(bearer_token=bearer_token)
tweets = client.search_recent_tweets(
    query="pineapple pizza",
    # tweet_fields=['created_at','lang'],
    # expansions="author_id",
    max_results=100
)
# print (tweets.data[0].created_at)
```

```python
df = pd.DataFrame([tweet.text for tweet in tweets.data], columns=["Tweets"])
df
```

```python
# Text sanitizing
def sanitize_text(text):
    text = re.sub(r"@[A-Za-z0-9]+", "", text) # Remove mentions
    text = re.sub(r"#", "", text) # Remove hashtag
    text = re.sub(r"RT[\s]+[: ]*", "", text) # Remove RT and any spaces after
    text = re.sub(r"https?:\/\/\S+", "", text) # Remove hyperlinks
    return text
df["Tweets"] = df["Tweets"].apply(sanitize_text)
```

```python
# Start tweet analysis

def get_subjectivity(text):
    return TextBlob(text).sentiment.subjectivity

def get_polarity(text):
    return TextBlob(text).sentiment.polarity

df["Subjectivity"] = df['Tweets'].apply(get_subjectivity)
df["Polarity"] = df['Tweets'].apply(get_polarity)
df
```

```python
# Visualize the vocabulary distribution

all_text = "".join(t for t in df["Tweets"])
word_cloud = WordCloud(width = 500, height = 500, random_state=15, max_font_size = 120).generate(all_text)

plt.imshow(word_cloud)
plt.axis("off")
plt.show()
```

```python
def polarity_judge(polarity):
    if polarity > 0:
        return "Positive"
    elif polarity == 0:
        return "Neutral"
    else:
        return "Negative"
```

```python
df['Analysis'] = df['Polarity'].apply(polarity_judge)
```

```python
# Plot Polarity and Subjectivity
plt.figure(figsize=(8,6))
for i in range(df.shape[0]):
    plt.scatter(df['Polarity'][i], df['Subjectivity'][i])

plt.title('Polarity against Sentiment')
plt.xlabel('Polarity')
plt.ylabel('Sentiment')
```

```python
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Counts")
df["Analysis"].value_counts().plot(kind="bar")
plt.show()
df["Analysis"].value_counts()
```
